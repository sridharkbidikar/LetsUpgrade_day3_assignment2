```
marks=73
if marks>=90:
  print ("you have secured A grade")
elif marks>=75 and marks<90:
  print ("you have secured B grade")
elif marks>=60 and marks<75:
  print ("you have secured C grade")
elif marks>=45 and marks<60:
  print ("you have secured D grade")
elif marks>=35 and marks<45:
  print ("you have secured E grade")
else:
  print ("sorry you have failed")
```

```
 
for i in range(1,1001):
 
  if i>0:
 
    for j in range(2,i):
 
        if(i % j==0):
 
            break
 
    else:
 
        print(i)
```

```
# python program to print prime numbers, range set by user
```

```
x=int(input("please enter a last number: ")) 
a=0
while (a<=x):
    count = 0
    for i in range(2, (a//2 + 1)):
        if(a % i == 0):
            count = count + 1
            break
 
    if (count == 0 and a != 1):
        print(a)
    a=a+1
```
